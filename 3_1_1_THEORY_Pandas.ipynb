## 3.9 Apply and lambda functions

```python
df_1 = pd.DataFrame(
    {
        'name': ['Peter', 'Juan', 'Melisa', 'Ana', 'Charles', 'Maria', 'Sonia', 'Peter', 'Melisa', 'Ana', 'Ana'],
        'age': [23, 20, 34, 40, 45, 21, 67, 55, 89, 3, 14],
        'salary': [100000, 120000, 130000, 140000, 150000, 160000, 170000, 180000, 190000, 200000, 210000]
    }
)
df_1
```

```python
df_1['age_plus_10'] = df_1['age'] + 10
df_1
```

```python

df_1['age_plus_10'] = df_1['age'].apply(lambda x: x + 10)
df_1

# More complex lambda function with multiple columns
df_1['age_plus_salary'] = df_1.apply(lambda x: x['age'] + x['salary'], axis=1)
df_1
```

## 3.10. Working with datetime

```python
df = pd.DataFrame(
    {
        'date': ['2-10-2026 18:30:0', '7-1-2021 9:45:30', '11-12-2015 4:55:1'],
        'value': [2, 3, 4]
    }
)

# See See strftime documentation for more information on format choices:
# https://docs.python.org/3/library/datetime.html#strftime-and-strptime-format-codes.
df['date'] = pd.to_datetime(df['date'], format="%d-%m-%Y %H:%M:%S")
display(df)
```

```python
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day']= df['date'].dt.day
display(df)
```

```python
# isocalendar() returns a 3-tuple containing ISO year, week number, and weekday
df['week_of_year'] = df['date'].dt.isocalendar().week
df['day_of_week'] = df['date'].dt.dayofweek
# A leap year is a year, which has 366 days (instead of 365)
df['is_leap_year'] = df['date'].dt.is_leap_year
display(df)
```

```python
mapping = {
    0: 'Monday',
    1: 'Tuesday',
    2: 'Wednesday',
    3: 'Thursday',
    4: 'Friday',
    5: 'Saturday',
    6: 'Sunday'
}
df['day_of_week_name'] = df['date'].dt.weekday.map(mapping)
display(df)
```

```python
today = pd.to_datetime('today')
df['age'] = today.year - df['date'].dt.year
display(df)
```

```python
condition = (df['date'] > '2016-7-1 9:45:30') & (df['date'] <= today)
display(df.loc[condition])
```

```python
condition = (df['date'] > '2016-7-1 9:45:30') & (df['date'] <= today)
display(df[condition])
```

## 3.11. Iterating over a pandas dataframe

```python
df = pd.DataFrame(
    {
        'name': ['Peter', 'Juan', 'Melisa', 'Ana', 'Charles', 'Maria', 'Sonia', 'Peter', 'Melisa', 'Ana', 'Ana'],
        'age': [23, 20, 34, 40, 45, 21, 67, 55, 89, 3, 14]
    }
)

amount_of_months = []
for idx, row in df.iterrows():
    amount_of_months.append(row['age'] * 12)

df['amount_of_months'] = amount_of_months
df
```
