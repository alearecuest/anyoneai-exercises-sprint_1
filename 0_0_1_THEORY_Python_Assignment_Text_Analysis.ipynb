How many people were registered each year?

```python
years = {}

with open("/content/registered-names-1922-2015.csv") as csvfile:
    reader = csv.DictReader(csvfile)

    count = 1000000

    for idx, row in enumerate(reader):

      # get year
      registration_year = int(row["year"])

      if registration_year in years:
        years[registration_year] += int(row["count"])

      else:
        years[registration_year] = int(row["count"])

      if idx == count:
        break

    print("People Reistered each year: ", years)

```

What is the shortest name?

```python
length = 9999
name = ''

with open("/content/registered-names-1922-2015.csv") as csvfile:
    reader = csv.DictReader(csvfile)

    count = 10000

    for idx, row in enumerate(reader):

      name_length = int(len(row["name"]))

      if name_length < length:
        name = row["name"]
        length = name_length

      if idx == count:
        break

    print(f"The shortest name is {name} with a length of {length}")


```

```python
import csv
from collections import defaultdict

FILEPATH = "/content/registered-names-1922-2015.csv"

def is_palindrome(text: str) -> bool:
    t = text.casefold()
    return t == t[::-1]

people_per_year = defaultdict(int)
total_per_name = defaultdict(int)

shortest_name = None
longest_name = None
palindrome_name = None

with open(FILEPATH, newline="", encoding="utf-8") as f:
    reader = csv.DictReader(f)
    for row in reader:
        name = row["name"]
        year = int(row["year"])
        count = int(row["count"])

        people_per_year[year] += count
        total_per_name[name] += count

        if shortest_name is None or len(name) < len(shortest_name):
            shortest_name = name

        if longest_name is None or len(name) > len(longest_name):
            longest_name = name

        if palindrome_name is None and is_palindrome(name):
            palindrome_name = name

people_per_year = dict(sorted(people_per_year.items()))

most_registered_name = max(total_per_name, key=total_per_name.get)
most_registered_count = total_per_name[most_registered_name]

print("People registered each year:")
for year, total in people_per_year.items():
    print(year, total)

print("Shortest name:", shortest_name)
print("Largest name:", longest_name)
print("Palindrome name:", palindrome_name if palindrome_name else "None")
print("Name with most registered people:", most_registered_name, most_registered_count)
```
